```python
import os
import shutil
```

```python
segs_path = "Z:/Datasets/MedicalImages/BrainData/SickKids/DIPG/segmentations" # some in folders, some post_bias_norm.nrrd
scans_path = "Z:/Datasets/MedicalImages/BrainData/SickKids/DIPG/exported_scans" # all of FLAIR.nrrd in patient id folders


inter_path = "C:/Users/Yina Gao/Documents/thesis-brain-tumour/data_output/dipg_before_preprocessing" 
# should have folders named after patient ids
# with 2 files: patient_id_FLAIR_post_bias_norm.nrrd and patient_id_FLAIR_post_bias_norm-label.nrrd
```

```python
# patient dirs input, might should have segmentation but might not have scan
patient_dirs = [path for path in os.listdir(segs_path) if os.path.isdir(os.path.join(segs_path, path))]
segs_nodir = [os.path.join(segs_path, path) for path in os.listdir(segs_path) if os.path.isfile(os.path.join(segs_path, path)) and path.endswith("norm-label.nrrd")]
# segs that are on their own (not in its own directory)
```

```python
# write segmentation files (no_dir)
for path in segs_nodir:
    patient_id = os.path.basename(path).split("_")[0]

    # make directory
    os.makedirs(os.path.join(inter_path, patient_id), exist_ok = True)
    shutil.copy(path, os.path.join(inter_path, patient_id, os.path.basename(path)))    
```

```python
# write segmentation files (had a dir), flair may or may not be missing

for patient in patient_dirs:
    os.makedirs(os.path.join(inter_path, patient), exist_ok=True)

    files = os.listdir(os.path.join(segs_path, patient))

    for f in files:
        # found segmentation file, copy it over
        if f.endswith('bias_norm-label.nrrd'):
            shutil.copy(os.path.join(segs_path, patient, f) ,os.path.join(inter_path, patient, f))

        # pre-existing flair file, copy it over too
        elif f.endswith('bias_norm.nrrd'):
            shutil.copy(os.path.join(segs_path, patient, f), os.path.join(inter_path, patient, f))
    
```

```python
# copy flairs over for remaining, if exists 
for dir in os.listdir(inter_path): # patient id
    flair_exists = False
    for f in os.listdir(os.path.join(inter_path, dir)):
        if f.endswith("bias_norm.nrrd"):
            flair_exists = True
            break
    
    if not flair_exists:
        print("scan previously missing for:", dir)
        src_path = os.path.join(scans_path, dir, "FLAIR.nrrd")

        if os.path.isfile(src_path):
            shutil.copy(src_path, os.path.join(inter_path, dir, f"{dir}_post_bias_norm.nrrd"))
        else:
            print("no flair file found for:", dir)
```
